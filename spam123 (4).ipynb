## Machine learning project of CREDIT CARD FRAUD DETECTION



# Importing libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing,
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Data Collection and preprocessing

```python
df = pd.read_csv('/kaggle/input/sms-spam-collection-dataset/spam.csv',encoding='latin-1')
```

```python
df.head()
```

```python
column_drop=['Unnamed: 2','Unnamed: 3','Unnamed: 4']
df.drop(columns=column_drop, inplace=True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
## replace the null values with a null string
mail_data = df.where((pd.notnull(df)),'')
```

```python
mail_data.head()
```

```python
mail_data.info()
```

```python
mail_data.shape
```

# label encoding

```python
## label spam mail as 0 ; and ham mail as 1
mail_data.loc[mail_data['v1']=='spam','v1',]=0
mail_data.loc[mail_data['v1']=='ham','v1',]=1
```

spam=0
ham=1

```python
#seperating the data as texts and labels
X = mail_data['v2']
Y = mail_data['v1']
```

```python
print(X)
```

```python
print(Y)
```

# spliting of data into train test

```python
X_train, X_test, Y_train, Y_test = train_test_split (X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

# Feature Extraction

```python
feature_extraction = TfidfVectorizer()

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_test_features)
```

```python
print(X_train_features)
```

# Training of Model

```python
## logistic regression
model= LogisticRegression()
```

```python
# training the logistic regression model  with training data
model.fit(X_train_features ,Y_train)
```

## Evaluting the training data

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, f1_score
```

```python
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
```

```python
# prediction on training data 
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ',accuracy_on_training_data)
```

```python
# prediction on testing data 
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on testing data : ',accuracy_on_test_data)
```

```python
# prediction on training data  calculating precision
prediction_on_training_data = model.predict(X_train_features)
precision_on_training_data = precision_score(Y_train, prediction_on_training_data)
```

```python
print('precision on training data : ',precision_on_training_data)
```

```python
# prediction on training data  calculating precision
prediction_on_test_data = model.predict(X_test_features)
precision_on_test_data = precision_score(Y_test, prediction_on_test_data)
```

```python
print('precision on testing data : ',precision_on_test_data)
```
